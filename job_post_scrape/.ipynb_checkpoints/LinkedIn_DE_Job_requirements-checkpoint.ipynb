```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
```

```python
driver = webdriver.Chrome(r'chromedriver')
```

```python
driver.get('https://www.linkedin.com/jobs/search/?f_E=2%2C3&f_JT=F&f_T=2732&geoId=102257491&keywords=data%20engineer&location=london%20england%20united%20kingdom&sortBy=R')

time.sleep(5) # Let the user actually see something!

page_text = driver.page_source

```

```python
soup = BeautifulSoup(page_text, 'html.parser')
```

```python
job_links = []
for job in soup.find_all('a'):
    try:
        if job['data-tracking-control-name'] == "public_jobs_jserp-result_search-card":
            job_links.append(job.get('href'))
    except:
        continue
```

```python
len(job_links)
```

```python
text_dump = []

for link in job_links:
    driver.get(link)
    time.sleep(3)
    page_text = driver.page_source
    soup = BeautifulSoup(page_text, 'html.parser')
    
    div_tags = soup.find_all('div') # All div tags on job description page
    
    # Observed that the description of a job had class 'description__text' 
    # Let's extract these
    relevant_tags = []
    for tag in div_tags:
        try:
    #         print(tag['class'])
            if "description__text" in tag['class']:
                relevant_tags.append(tag)
                print('success')
        except:
            continue
            
    # Extracting the human readable text from the relevant tags
            
    for i in relevant_tags:
        text_dump.append(i.get_text(separator="\n"))
```

```python
word_counts = {}


for para in text_dump: 
    individual_words = para.split(' ')

    for word in individual_words:
        if word not in word_counts:
            word_counts[word] = 1
        else:
            word_counts[word] += 1
        
    print('success')
```

```python
import pandas as pd

df_frequencies = pd.DataFrame.from_dict(word_counts, orient='index', columns=['freq']).sort_values('freq', ascending=False)
```
